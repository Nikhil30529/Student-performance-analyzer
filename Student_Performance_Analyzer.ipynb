```python
subject_columns = ['Maths','Physics','Chemistry','English','Biology','Economics','History','Civics']
student['Average_Score'] = student[subject_columns].mean(axis=1)
def assign_grade(score):
    if score >= 80:
        return 'A'
    elif score >= 60:
        return 'B'
    elif score >= 40:
        return 'C'
    else:
        return 'D'
student['Grade'] = student['Average_Score'].apply(assign_grade)
student[['Unnamed: 0', 'Average_Score', 'Grade']].head()
```
